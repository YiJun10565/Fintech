```python
import pandas as pd
import csv

ETF_list = pd.read_csv(
    "./data/DUST.csv",
    usecols=['Date'],
    parse_dates=True)

def add_pd(ETF_name):
    path = './data/'+ETF_name+'.csv'
    ETF_list[ETF_name] = pd.read_csv(
    filepath_or_buffer = path,
    usecols = ['Adj Close'],
    parse_dates=True
    )
    
add_pd('CUT')
add_pd('DUST')

ETF_list


```
